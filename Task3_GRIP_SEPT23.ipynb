### **Data Visualization**

```python
# Displaying the Correlation between Variables
store_data.corr(method='pearson')
```

```python
# Finding the pairwise correlations between the columns and visualizing using heatmap

sns.heatmap(store_data.corr(numeric_only=True),square=True)
```

```python
# Distribution of Orders on the basis of Shipping Mode
myexplode = [0.1, 0, 0, 0]
store_data['Ship Mode'].value_counts().plot(kind='pie',figsize=[10,10],explode=
myexplode,autopct='%1.1f%%')
plt.title(' Orders Distribution on the Basis of Mode of Shipping')
```

Observation: The 'Standard Class' shipping mode has the highest share of orders at 59.7%.

```python
# Orders Distribution on the Basis of Region
myexplode = [0.1, 0, 0, 0]
store_data['Region'].value_counts().plot(kind='pie',figsize=[10,10],
explode=myexplode,autopct='%1.1f%%')
plt.title(' Orders Distribution on the Basis of Region')
```

Observation: The 'West' region has the highest share of orders at 32% .

```python
# Orders Distribution on the Basis of Category
myexplode = [0.1, 0, 0]
store_data['Category'].value_counts().plot(kind='pie',figsize=[10,10],
explode=myexplode,autopct='%1.1f%%')
plt.title('Orders Distribution on the Basis of Category')
```

Observation: The 'Office Supplies' category has the highest share of orders at 60.3% .

```python
x=store_data.Category
y=store_data.Sales
fig = plt.figure(figsize = (10, 5))
plt.bar(x,y)
plt.title("Category-Wise Sales Distribution")
plt.xlabel("Category",fontweight ='bold', fontsize = 13)
plt.ylabel("Sales",fontweight ='bold', fontsize = 13)
plt.show()
```

```python
# Bar chart showing Region-Wise Sales Distribution
x=store_data.Region
y=store_data.Sales
# setting figure size by using figure() function
fig = plt.figure(figsize = (10, 5))
plt.bar(x,y)
plt.title("Region-Wise Sales Distribution")
plt.xlabel("Region",fontweight ='bold', fontsize = 13)
plt.ylabel("Sales",fontweight ='bold', fontsize = 13)
plt.show()
```

```python
# Bar chart showing Segment-Wise Sales Distribution
x=store_data.Segment
y=store_data.Sales
# setting figure size by using figure() function
fig = plt.figure(figsize = (10, 5))
plt.bar(x,y)
plt.title("Segment-Wise Sales Distribution")
plt.xlabel("Segment",fontweight ='bold', fontsize = 13)
plt.ylabel("Sales",fontweight ='bold', fontsize = 13)
plt.show()
```

Observation: The highest Sales is offered by the Home Office segment for the Technology products in the Southern region. So, the company should come up with attractive deals to maximise profit.

```python
# visualizing the whole dataset using the pair plot
import seaborn as sns
sns.pairplot(store_data,hue='Region')
plt.show
```

```python
store_data['City'].value_counts().head(20).plot(kind = 'barh',color='hotpink')
plt.title('Top 20 Cities having the most orders placed')
```

Observation:Long Beach has the lowest number of orders placed therefore, find out exactly what is causing less ordering trend there.Is it due to lack of motivation of the Sales team? (OR) Is it due to the pricing model of products as compared to other competitors in the market?

```python
store_data['State'].value_counts().head()
```

Observation: Maximum orders are placed in California followed by New York due to the concentration of higher income family groups there and higher disposable incomes. This also shows the highest levels of customer satisfaction in these areas.

```python
df.plot(kind='bar',figsize=(15,5),color='orange')
plt.ylabel('Number of Orders')
plt.xlabel('City')
plt.title('City Wise Order Distribution', fontweight='bold', fontsize = 18)
plt.show()
```

```python
segment_wise= store_data.groupby("Segment").sum(numeric_only= True)
segment_wise
```

```python
segment_wise[['Sales', 'Profit']].plot.bar()
plt.xlabel("Segment",fontsize=13)
plt.ylabel("Total Sales or Total Profit",fontsize=13)
plt.title("Sales and Profit Distribution over Segment",fontweight='bold',fontsize=15)
```

```python
category_wise=store_data.groupby("Category").sum(numeric_only=True)
category_wise
```

```python
category_wise[['Sales', 'Profit']].plot.bar()
plt.xlabel("Category",fontsize=13)
plt.ylabel("Total Sales or Total Profit",fontsize=13)
plt.title("Sales and Profit Distribution over Category",fontweight='bold',fontsize=15)
```

```python
subcategory_wise=store_data.groupby("Sub-Category").sum(numeric_only=True)
subcategory_wise
```

```python
subcategory_wise[['Sales', 'Profit']].plot.bar()
plt.xlabel("Sub-Category",fontsize=13)
plt.ylabel("Total Sales or Total Profit",fontsize=13)
plt.title("Sales and Profit Distribution over Sub-Category",fontsize=15,fontweight='bold')
```

Observation: The maximum Sales and Profit has been offered by the Consumer segment in the Technology products category. Thus, the company can generate long-term cash flow and boost revenue growth through leveraging it's influence on this section of it's customer base.

```python
sns.barplot(x='Discount',y='Profit',data=store_data,palette ='plasma')
plt.xlabel("Discount",fontsize=13)
plt.ylabel("Profit",fontsize=13)
plt.title("Profit Vs Discount",fontsize=15,fontweight='bold')
```

From the Profit Vs Discount barplot, it can be observed that the company generated maximum profit when the discount rates were between 8% to 12%.

```python
sns.barplot(x='Discount',y='Sales',data=store_data,palette ='plasma')
plt.xlabel("Discount",fontsize=13)
plt.ylabel("Sales",fontsize=13)
plt.title("Sales Vs Discount",fontsize=15,fontweight='bold')
```

Observation: The maximum Sales is generated when the discount rate is around 50%. But the total profit is the lowest at 50% discount rate.Therefore, a trade-off has to be made between Total Sales, Discount Rate and Total Profit.

```python
state_wise= store_data.groupby("State").sum(numeric_only=True)
state_wise
```

```python
state_wise['Profit'].plot(kind='bar',figsize=[20,8],color='orange')
plt.xlabel("State",fontweight='bold',fontsize=13)
plt.ylabel("Profit",fontweight='bold',fontsize=13)
plt.title("State-Wise Profit Distribution",fontsize=15,fontweight='bold')
plt.show()
```

Observation: Since, Profit and Sales have a direct positive correlation with each other therefore, it can be observed that the Top 2 Cities and States with the highest Total Profit and Total Sales are California and New York.

### **Conclusions from the Exploratory Data Analysis (EDA) of 'SampleSuperstore' Dataset:**

1) In New York and California, the sales and marketing team should put in effort to maximise profit and sales through running promotional campaigns to introduce the customer base to new products and conduct a competitor analysis to ascertain whether customers consider the company's price point to be good value.  

2) We should increase the sales of Copiers as the difference between Total Sales and Total Profit for Copiers is low and the Sales and Profit levels are also good as compared to other sub-categories.

 3) We should keep the discount rate from 8% to 12 % as within this range the Total Sales and Total Profit are maximised.

 4) According to the barplot for Sales & Profit distribution over Sub-Categories, there is a huge loss on the sales of tables. So, its sales should be minimized to reduce operating costs. We should focus on branding and quality which can help sustain higher prices on sales and ensure higher profit margins over the long term.
