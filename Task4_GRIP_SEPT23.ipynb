### **Data Preparation & Cleaning**

```python
df.isnull().sum() # Checking for null values in the dataset
```

```python
df.duplicated().sum() # Checking for duplicate values
```

```python
# Displaying the columns in the dataset
df.columns
```

```python
# Renaming the columns
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country',
'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group',
'targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
df=df[['Year','Month','Day','Country','Region','city','latitude','longitude','AttackType',
'Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
# Replacing all NaN elements in the "Killed" column with 0s
df["Killed"]=df["Killed"].fillna(0)
# Replacing all NaN elements in the "Wounded" column  with 0s
df["Wounded"]=df["Wounded"].fillna(0)
# Concatinating the columns "Killed" and "Wounded"
df["Casualty"]=df["Killed"]+df["Wounded"]
df.head(10)
```

###**Data Visualization**

```python
print("Country with the most frequent attacks:",df['Country'].value_counts().idxmax())
# Omitting instances of Unknown in the column "City"
print("Most frequently attacked city:",df['city'].value_counts().index[1])
print("Most Common Attack Type:",df['AttackType'].value_counts().idxmax())
print("Region with the most attacks:",df['Region'].value_counts().idxmax())
print("Year with the most number of attacks:",df['Year'].value_counts().idxmax())
print("Month with the most number attacks:",df['Month'].value_counts().idxmax())
# Omitting instances of Unknown in the column "Group"
print("Group with the most number attacks:",df['Group'].value_counts().index[1])
```

```python
# Year with most number of attacks is 2014
df['Year'].value_counts(dropna = False).sort_index()
```

```python
x_year = df['Year'].unique() # Taking only unique "Year" values
y_count_years = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'rocket')
plt.xticks(rotation = 45) # Rotating x axis values by 45 degree
plt.xlabel('Attack Year',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.title('Number of Attacks on the Basis of Year of Occurrence',fontweight='bold',fontsize=18)
plt.show()
```

```python
df['Region'].value_counts()
```

```python
df['Region'].value_counts()[:6].plot(kind='pie',figsize=[20,10],autopct='%1.1f%%')
plt.title('Region Wise Distribution of Attacks (Top 6)',fontsize=18,fontweight='bold')
```

```python
df['AttackType'].value_counts()[:5].plot(kind='pie',figsize=[20,10],autopct='%1.1f%%')
plt.title('Distribution of Attacks on the Basis of Type of Attack (Top 5)',fontsize=18,fontweight='bold')
```

```python
df['Target_type'].value_counts()[:5].plot(kind='pie',figsize=[20,10],autopct='%1.1f%%')
plt.title('Distribution of Attacks on the Basis of Target Type (Top 5)',fontsize=18,fontweight='bold')
```

```python
x= df['Country'].value_counts()[:5].index
y=df['Country'].value_counts()[:5]
plt.bar(x,y,color='orange')
plt.xlabel('Countries',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.title("Hot Zones of Terrorism (Top 5)",fontweight='bold',fontsize=18)
plt.show()
```

```python
x= df['city'].value_counts()[1:6].index
y=df['city'].value_counts()[1:6]
plt.bar(x,y,color='orange')
plt.xlabel('Cities',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.title("Top 5 Affected Cities by Terrorism",fontweight='bold',fontsize=18)
plt.show()
```

```python
Country_Wise_Casualties= df.groupby("Country")['Killed','Wounded','Casualty'].sum(numeric_only= True).sort_values('Casualty',ascending=False)
Country_Wise_Casualties
```

```python
df[['AttackType','Casualty']].groupby(["AttackType"],axis=0).sum().plot(kind='bar',figsize=(20,10),color='orange')
plt.xticks(rotation=50)
plt.title("Casulaties on the basis of Type of Attack",fontsize=15)
plt.ylabel('Casualty',fontsize=13)
plt.xlabel('Attack type',fontsize=13)
plt.show()
```

```python
data=df[['Group','Country','Casualty']]
Terrorist_Group=data.groupby(['Group','Country'],axis=0).sum().sort_values('Casualty',ascending=False).drop('Unknown').head(10)
Terrorist_Group
```

### **Conclusions from the Exploratory Data Analysis (EDA) of 'Global Terrorism' Dataset**

1) Iraq and Pakistan are the top 2 hot zones of terrorism.

2) Baghdad and Karachi are the two cities most affacted by terrorist attacks.

3) Year of highest terrorist attacks: 2014

4) Most common type of terrorist attack: Bombing/Explosion.

5) Region with the most terror attacks: Middle East & North Africa

6) Month with the most number of attacks: 5 (i.e. the month of May)

7) Group with the most number of attacks: Taliban

8) Highest casualties are caused by Bombing/explosion.

9) The most attacked target type is the private citizens and property.

10) Country-wise Iraq has the highest number of casualties.

11) Taliban of Afghanistan has caused the highest number of casualties.
